```python
import pandas as pd
from bertopic import BERTopic
import os

# Configuration
DATA_PATH = "../data/fed_minutes_sentences_structured.csv"
```

## 1. Load Data Minutes

```python
if os.path.exists(DATA_PATH):
    df = pd.read_csv(DATA_PATH)
    # Filter for meaningful sentences (e.g., > 5 words)
    df = df[df['sentence_text'].str.split().str.len() > 5]
    print(f"Loaded {len(df)} sentences for modeling.")
    display(df.head())
else:
    print(f"File not found: {DATA_PATH}")
```

## 2. Train BERTopic Model

We will train a BERTopic model on the sentences. Note: This may take some time depending on the dataset size and hardware.

```python
if 'sentence' in df.columns:
    docs = df['sentence'].tolist()
    
    # Initialize and fit the model
    topic_model = BERTopic(language="english", calculate_probabilities=False, verbose=True)
    topics, probs = topic_model.fit_transform(docs)
    
    # Get topic info
    freq = topic_model.get_topic_info()
    display(freq.head(10))
```

## 3. Visualize Topics

Visualizing the intertopic distance and topic hierarchy.

```python
topic_model.visualize_topics()
```
